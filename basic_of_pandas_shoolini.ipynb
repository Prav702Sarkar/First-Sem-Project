You can import Pandas with import pandas as pd.

# **Let's explore the unique and missing values**

* ### <code>unique()</code>
* ### <code>nunique()</code>
* ### <code>value_counts()</code>
* ### <code>isnull()</code>
* ### <code>fillna()</code>

```python
#find the unique values in python
df["Sex"].unique()
```

```python
#find the number of unique value in column
df["Sex"].nunique()
```

```python
#count the number of unique value in a column
df["Sex"].value_counts()
```

```python
#find the number of null value in a column
df.isnull().sum()
```

```python
#lets fill the missing value in a column
df["Age"].fillna(df["Age"].mean(), inplace = True)
```
